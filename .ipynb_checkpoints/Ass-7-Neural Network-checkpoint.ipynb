```python
import scipy.io
import numpy as np
import pandas as pd
```

```python
f = scipy.io.loadmat('ex4data1.mat')
data=f['X']
out=f['y']
x,y=data.shape
out.shape
```

```python
one=np.ones((x,1))
one
data1=np.hstack((data,one))
x1,y1=data1.shape
data1[:5,:]
```

```python
W1=np.random.randn(26,y1)
W1.shape
W1[:5,:]
```

```python
data2=np.dot(W1,data1.T)
data2.shape
data2[:5,:]
```

```python
def sigmoid(data):
    return 1/(1+np.exp(-1*data))
```

```python
def maximizer(data):
    return np.argmax(data,axis=1)
```

```python
act_data2=sigmoid(data2)
x3,y3=act_data2.shape
act_data2[:5,:]
```

```python
# one2=np.ones((1,x1))
# act_data2=np.vstack((act_data2,one2))
```

```python
W2=np.random.randn(10,x3)
W2.shape
```

```python
data3=np.dot(W2,act_data2)
act_data3=sigmoid(data3).T
act_data3.shape
act_data3[:5,:]
```

```python
prediction_final=maximizer(act_data3)
prediction_final.shape
```

```python
data_X=np.zeros((x1,10))
print(data_X.shape)
for i,j in zip(out,range(0,x1)):
    data_X[j,i%10]=1
data_X
```

```python
alpha=0.01
loss=1/x1*np.sum(np.sum((act_data3-data_X)**2,axis=1),axis=0)
loss
```

```python
dz2=act_data3-data_X
print(dz.shape)
for i in range(0,10):
    if i==0:
         dw=np.sum(dz[:,i].T*act_data2,axis=1)
    else:
        x=np.sum(dz[:,i].T*act_data2,axis=1)
        dw=np.vstack((dw,x))
W2=W2-alpha*1/x1*dw
W2
```

```python
dz1=np.dot(W2.T,dz2.T)*((act_data2*(1-act_data2)))
dw1=np.dot((np.dot(W2.T,dz2.T)*(act_data2*(1-act_data2))),data1)
W1=W1-alpha*1/x1*dw1
W1
```
